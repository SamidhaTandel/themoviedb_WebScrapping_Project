```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.themoviedb.org/movie'  #put your url in this quote
header = {'User-Agent': 'Mozilla/5.0 (Windows NT 6.3; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/99.0.4844.51 Safari/537.36'}
source_data = requests.get(url, headers = header).text

data_beautify=BeautifulSoup(source_data,'lxml')
all_data=data_beautify.find('div',class_='card style_1')
data2=data_beautify.find_all('p',class_='load_more')[1]

all_links = [item['href'] for item in data2]

# Convert the list to a single string with each link separated by a newline character
links_string = '\n'.join(all_links)

page_link=links_string[:-1]
print(page_link)




url2=[]
for i in range(1,51):
    url2.append("https://www.themoviedb.org"+page_link+str(i))
    
print(url2)

Movie_list=[]
for i1 in url2:
    header = {'User-Agent': 'Mozilla/5.0 (Windows NT 6.3; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/99.0.4844.51 Safari/537.36'}
    source_data = requests.get(i1, headers = header).text
    data_beautify=BeautifulSoup(source_data,'lxml')
    all_data=data_beautify.find_all('div',class_='card style_1')
    
    for item in all_data:
        movie_name=item.find('h2').text
        links=item.select('a')
        actual_web_links = [web_link['href'] for web_link in links]
        link=[]
        
        for i in actual_web_links:
            link.append('https://www.themoviedb.org'+str(i))
        
        all_link=link[0]
        header = {'User-Agent': 'Mozilla/5.0 (Windows NT 6.3; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/99.0.4844.51 Safari/537.36'}
        data = requests.get(all_link, headers = header).text
        
        new_data=BeautifulSoup(data,'lxml')
        
        all_info=new_data.find('div',class_='header_poster_wrapper')
        
        rating=all_info.find('div',class_='user_score_chart')
        
        if rating:
            rate=rating['data-percent']
            
            
        else:
            'Data not found'
            
        profile=all_info.find('ol',class_='people no_image')
#         director=all_info.find_all('ol')[0].text.replace("\nDirector, Screenplay, Story\n\n\nRoald Dahl\nCharacters\n\n\nSimon Farnaby\nScreenplay\n\n","").replace("\n","")
        directors=all_info.find('li',class_='profile')
        if directors:
            dire = directors.find('p').text
        else:
            dire = 'N/A'

        screenplay=all_info.find_all('li')[2].text.replace("\n","").replace("Screenplay","")
        
        movie_info=all_info.find('div',class_='facts')
        
        
        duration=movie_info.find('span',class_='runtime').text.replace("\n","") if movie_info.find(class_='runtime')else 'Duration name not found'

        release_date=movie_info.find('span',class_='release').text.replace("\n","").replace("(IN)","").strip()
        genres=movie_info.find('span',class_='genres').text.replace("\n","").replace(',\xa0'," ")
        
        all_movie={
                "Movie Name": movie_name,
                "Release Date":release_date,
                "Director":dire,
                "Duration":duration,
                "Genre":genres,
                "Rating(Out of 100%)":rate
            
            }
        Movie_list.append(all_movie)
```

```python
import pandas as pd
# pip install openpyxl
df=pd.DataFrame(Movie_list)
df.to_excel('Movie_details_scrapping.xlsx')
df
```
